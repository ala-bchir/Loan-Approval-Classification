```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import make_scorer, accuracy_score, recall_score, precision_score, confusion_matrix, ConfusionMatrixDisplay
from sklearn.metrics import classification_report
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import export_graphviz
import graphviz
```

```python
df = pd.read_csv('../Data/cleaned_loan_data.csv')
```

### Handling Target feature Imbalance
To balance the dataset so that the model doesnâ€™t become biased toward the majority class (`loan_status = 0`) we use : 
1. The **`stratify`** argument of `train_test_split` ensures that each class is proportionally represented in both the training and test sets.
2.  SMOTE-Synthetic Minority Over-sampling Technique : generates synthetic samples for the minority class instead of simply duplicating existing data points.

```python
X = df.iloc[:,:-1].values
Y = df.iloc[:,-1].values.reshape(-1,1)
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=.2,random_state=42,stratify=Y)
```

```python
smote = SMOTE(random_state=42)
X_train_resampled, Y_train_resampled = smote.fit_resample(X_train, Y_train)
```

```python
print(pd.Series(Y_train.flatten()).value_counts())
print(pd.Series(Y_train_resampled.flatten()).value_counts())
```

# Tuning

## Grid search with cross-validation

```python
def grid_search (X_train, Y_train) :
    tree = DecisionTreeClassifier(random_state=42)
    
    param_grid = {
        "max_depth": [2,3, 5, 7, 9,12], 
        "min_samples_split": [2, 5, 10, 20,100], 
        # "class_weight": [None, "balanced"], 
        # "max_features": [None, "sqrt", "log2", 0.5, 0.8], 
        # "max_leaf_nodes": [None, 10, 20, 50, 100],
        # "min_weight_fraction_leaf": [0.0, 0.01, 0.05, 0.1], 
        # "min_impurity_decrease": [0.0, 0.001, 0.01, 0.1]
    }
    scoring = {
        'accuracy': 'accuracy',
        'recall': make_scorer(recall_score),
        'precision': make_scorer(precision_score)
    }
    
    grid_search = GridSearchCV(
        estimator=tree,
        param_grid=param_grid,
        cv=5,  
        scoring=scoring, 
        refit='accuracy',  
        n_jobs=-1,
        verbose=1
    )
    grid_search.fit(X_train, Y_train.ravel())  

    # Best hyperparameters
    best_params = grid_search.best_params_
    print("Best parameters :", best_params)
    print("Best accuracy : ", grid_search.best_score_)
```

# Models

## Decision Tree Model

```python
def Decision_Tree_Model (max_depth, min_samples, X_train, Y_train) : 

    final_tree = DecisionTreeClassifier(min_samples_split = min_samples, max_depth= max_depth, random_state=42)
    final_tree.fit(X_train, Y_train.ravel())
    
    Y_pred = final_tree.predict(X_test)
    accuracy = accuracy_score(Y_test, Y_pred)
    recall = recall_score(Y_test, Y_pred) 
    precision = precision_score(Y_test, Y_pred)
    #f1 = f1_score(Y_test, Y_pred)
    print(f"Decision Tree Model Performance:")
    print(f" - Accuracy: {accuracy:.4f}")
    print(f" - Recall: {recall:.4f}")
    print(f" - Precision: {precision:.4f}")
    #print("F1 score :", f1)
    return Y_test, Y_pred, final_tree
```

## Random Forest Model

```python
def random_forest(X_train, Y_train, X_test, Y_test): 
    # Initialize and train the Random Forest model
    rf_model = RandomForestClassifier(
        n_estimators=100,       
        max_depth=12,          
        min_samples_split=2,    
        random_state=42
    )
    rf_model.fit(X_train, Y_train.ravel())
    
    # Predict on the test set
    Y_pred_rf = rf_model.predict(X_test)
    
    # Evaluate performance
    accuracy_rf = accuracy_score(Y_test, Y_pred_rf)
    recall_rf = recall_score(Y_test, Y_pred_rf)
    precision_rf = precision_score(Y_test, Y_pred_rf)
    
    # Display metrics
    print(f"Random Forest Model Performance:")
    print(f" - Accuracy: {accuracy_rf:.4f}")
    print(f" - Recall: {recall_rf:.4f}")
    print(f" - Precision: {precision_rf:.4f}")
    
    # Return relevant outputs
    return Y_test, Y_pred_rf, rf_model
```

```python
#To display the results
def plot_confusion_matrix(title, Y_test, Y_pred):
    cm = confusion_matrix(Y_test, Y_pred)
    # Plot confusion matrix
    plt.figure(figsize=(6, 5))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=["Refused", "Approved"], 
                yticklabels=["Refused", "Approved"])
    plt.xlabel("Predicted Values")
    plt.ylabel("Actual Values")
    plt.title(title) 
    plt.show()
```

# Another Implementation of the Decision Tree Algorithm ...

```python
class Noeud():
    def __init__(self,seuil=None,feature_index=None,gain_info=None,right=None,left=None,valeur=None):
        self.seuil = seuil
        self.feature_index = feature_index
        self.gain_info = gain_info
        self.right = right
        self.left = left
        self.valeur = valeur #que pour les feuilles

class Arbre_Decision():
    def __init__(self,min_samples_split=2,prof_max=3):
        
        self.root = None
        self.min_samples_split = min_samples_split
        self.prof_max = prof_max
        
    def build_tree(self,dataset,prof=0):
        """fct recursive de construction d'arbre"""

        X,Y = dataset[:,:-1], dataset[:,-1]
        num_samples , num_features = np.shape(X)

        if num_samples>=self.min_samples_split and prof <= self.prof_max:
            best_split = self.best_split(dataset,num_samples,num_features)
            if best_split["gain_info"] > 0 :
                left = self.build_tree(best_split["dataset_left"],prof+1)
                right = self.build_tree(best_split["dataset_right"],prof+1)

                return Noeud(best_split["seuil"],best_split["feature_index"],best_split["gain_info"],right,left)
        valeur_feuille = self.valeur_feuille(Y)
        return Noeud(valeur = valeur_feuille)

    def split(self,dataset,feature_index,seuil):
        '''separation des donnÃ©es par rapport a notre seuil'''
        dataset_left = np.array([row for row in dataset if row[feature_index]<= seuil])
        dataset_right = np.array([row for row in dataset if row[feature_index]> seuil])
        return dataset_left,dataset_right

    def best_split(self,dataset,num_samples,num_features):
        '''trouver la meilleure separation'''
        best_split=dict()
        max_gain = -float("inf")
        
        for feature_index in range(num_features):
            feature_values = dataset[:,feature_index]
            seuils_possible = np.unique(feature_values)

            for seuil in seuils_possible:
                #separation 
                dataset_left,dataset_right = self.split(dataset,feature_index,seuil)
                # verification des fils (non null)
                if len(dataset_left)>0 and len(dataset_right)>0:
                    y,left_y,right_y = dataset[:, -1] , dataset_left[:, -1] , dataset_right[:, -1]
                    #calculer le gain d'information de la separation courante
                    curr_gain = self.gain_information(y,right_y,left_y)
                    if curr_gain > max_gain:
                        #on est met a jour notre dict
                        best_split["feature_index"] = feature_index
                        best_split["seuil"]         = seuil
                        best_split["dataset_right"] = dataset_right
                        best_split["dataset_left"]  = dataset_left
                        best_split["gain_info"]     = curr_gain
                        
                        max_gain = curr_gain
        return best_split


    def gini(self,y):
        '''calculer l'index gini'''
        labels = np.unique(y)
        gini = 0 
        for l in labels:
            pro_l = len(y[y==l]) / len(y)
            gini += pro_l**2
        return 1-gini

    def entropy(self,y):
        '''calculer l'entropy'''
        labels = np.unique(y)
        entropy = 0 
        for l in labels:
            pro_l = len(y[y==l]) /len(y)
            entropy += -pro_l * np.log2(pro_l)
        return entropy

    def gain_information(self,parent,r_child,l_child,m="gini"):
        '''calculer le gain d'information'''
        weight_r = len(r_child) / len(parent)
        weight_l = len(l_child) / len(parent)

        if m == "gini":
            gain = self.gini(parent) - (weight_r*self.gini(r_child) + weight_l*self.gini(l_child))
        else:
            gain = self.entropy(parent) - (weight_r*self.entropy(r_child) + weight_l*self.entropy(l_child))
        return gain
 
    def valeur_feuille(self,y):
        '''calculer l'appartenance de la feuille'''
        y = list(y)
        return max(y,key=y.count)
    
    def fit(self,X,Y):
        dataset = np.concatenate((X,Y),axis=1)
        self.root = self.build_tree(dataset)

    def single_pred(self,x,noeud):

        if noeud.valeur != None: #feuille
            return noeud.valeur
        feature_val = x[noeud.feature_index]
        if feature_val <= noeud.seuil:
            return self.single_pred(x,noeud.left)
        else:
            return self.single_pred(x,noeud.right)

    def predict(self,X):
        preds = [self.single_pred(x,self.root) for x in X]
        return preds
```

```python
classifier = Arbre_Decision(min_samples_split=20,prof_max=9)
classifier.fit(X_train,Y_train)
```

```python
Y_pred = classifier.predict(X_test)
accuracy = accuracy_score(Y_test, Y_pred)
print("PrÃ©cision sur l'ensemble de test :", accuracy)
```
